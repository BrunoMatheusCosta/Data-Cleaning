```python
# Importando bibliotecas

import pandas as pd
```

```python
# Importando base de dados

df = pd.read_csv('fifa21_raw_data.csv')
df.head()
```

```python
# Usando a função info para verificar valores faltantes e nome das colunas

df.info()
```

```python
# Usaremos apenas algumas colunas do DataSet

df = df[['Team & Contract', 'Hits', 'Joined', 'Height', 'Weight', 'Value', 'Wage', 'Positions']]
df.head()
```

```python
# Removendo caracteres indesejados das variáveis (coluna 'Team & Contract' sera tratada posteriormente)

df['Hits'] = df['Hits'].str.replace('\n','')
df['Weight'] = df['Weight'].str.replace('lbs','')
df['Value'] = df['Value'].str.replace('€','')
df['Wage'] = df['Wage'].str.replace('€','')
df.head()
```

```python
# Tratando coluna 'Team & Contract' (Separando clube do contrato e removendo caracteres indesejados)

df['Team & Contract'] = df['Team & Contract'].astype('str') 
df['Team & Contract'].replace('\n', '', regex=True, inplace=True)
test = df['Team & Contract'][0]
Team = []
Contract_Duration = []
for x in range(len(df['Team & Contract'])):
    Team.append(str(df['Team & Contract'][x][:-11]))
    c = str(df['Team & Contract'][x][-11:])
    if c.startswith("2") == True: 
        Contract_Duration.append(c)
    else:
        Contract_Duration.append("0")
df = df.drop(columns = ['Team & Contract'])   
df['Team'] = Team
df['Contract Duration'] = Contract_Duration
df.head()
```

```python
# Convertendo Height (Altura) e Weight (Peso) para o padrão internacional de medidas (cm e kg)

temp_height = []
for x in range(len(df['Height'])):
    x = df['Height'][x].replace('"', '').split("'")
    inches = int(x[0]) * 12 + int(x[1])
    temp_height.append(inches)
df['Height'] = temp_height
df['Height'] = df['Height'].astype('int64')
df['Height'] = df['Height'] * 2.54

df['Weight'] = df['Weight'].astype('int64')
df['Weight'] = df['Weight'] / 2.205

df.head()
```

```python
# Definindo função que converterá o mês categorico para numerico

def month_to_number(x):
            if x== 'Jan':
                return 1
            elif x == 'Feb':
                return 2
            elif x == 'Mar': 
                return 3
            elif x == 'Apr':
                return 4
            elif x == 'May':
                return 5
            elif x == 'Jun': 
                return 6
            elif x == 'Jul': 
                return 7
            elif x == 'Aug': 
                return 8
            elif x == 'Sep': 
                return 9
            elif x == 'Oct': 
                return 10
            elif x == 'Nov':  
                return 11
            elif x == 'Dec':  
                return 12
```

```python
# Convertendo a coluna 'Joined' para o formato DateTime

date = []
for x in range(len(df['Joined'])):
    d = df['Joined'][x]
    c = d.split(" ")
    month = str(month_to_number(c[0]))
    day = str(c[1].replace(',',''))
    year = str(c[2])
    if len(day) == 1:
        day = ('0'+str(day))
    date_long = (str(month)+'/'+str(day)+'/'+str(year))
    date_con = pd.to_datetime(date_long)
    date.append(date_con)  
df['Joined'] = date
df.head()
```

```python
# Convertendo as colunas 'Value' e 'Wage' para formato totalmente numerico

def value_to_float(x):
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0

df['Value'] = df['Value'].apply(value_to_float)
df['Wage'] = df['Wage'].apply(value_to_float)
df.head()
```

```python
# Traduzindo colunas para o português

Traducao = {
    'Hits' : 'Acertos',
    'Joined' : 'Entrada',
    'Height' : 'Altura',
    'Value' : 'Valor',
    'Wage' : 'Salario',
    'Positions' : 'Posicoes',
    'Team' : 'Clube',
    'Contract Duration' : 'Contrato',
}

df = df.rename(columns=Traducao)
df.head()
```
